DATA UNDERSTANDING

This data contains information about people involved in a crash and if any injuries were sustained. My analysis levarages the traffic crashes dataset that contains over 1,840,933 records and 29 distinct features(columns). With a time period from November 2015 to May 2024,each record corresponds to an occupant in a vehicle listed in the Crash dataset. Some people involved in a crash may not have been an occupant in a motor vehicle, but may have been a pedestrian, bicyclist, or using another non-motor vehicle mode of transportation. Injuries reported are reported by the responding police officer. Fatalities that occur after the initial reports are typically updated in these records up to 30 days after the date of the crash. Person data can be linked with the Crash and Vehicle dataset using the “CRASH_RECORD_ID” field. A vehicle can have multiple occupants and hence have a one to many relationship between Vehicle and Person dataset. However, a pedestrian is a “unit” by itself and have a one to one relationship between the Vehicle and Person table.

The Traffic crashes dataset contains the following columns:

1.PERSON_ID: A unique identifier for each individual involved in the accident. This could be a driver, passenger, pedestrian, or cyclist.

2.PERSON_TYPE: The role of the person in the accident, such as driver, passenger, pedestrian, or cyclist.

3.CRASH_RECORD_ID: A unique identifier for each accident record.

4.VEHICLE_ID: A unique identifier for each vehicle involved in the accident.

5.CRASH_DATE: The date and time when the accident occurred.

6.SEAT_NO: The seat position of the person in the vehicle (e.g., driver seat, front passenger seat, rear seats).

7.CITY: The city where the accident took place.

8.STATE: The state where the accident occurred.

9.ZIPCODE: The postal code of the location where the accident happened.

10.SEX: The gender of the person involved in the accident.

11.AGE: The age of the person involved in the accident.

12.DRIVERS_LICENSE_STATE: The state that issued the driver’s license of the driver involved in the accident.

13.DRIVERS_LICENSE_CLASS: The class of the driver’s license (e.g., commercial, non-commercial).

14.SAFETY_EQUIPMENT: The type of safety equipment used by the person (e.g., seat belt, helmet).

15.AIRBAG_DEPLOYED: Indicates whether the airbag was deployed during the accident.

16.EJECTION: Indicates whether the person was ejected from the vehicle during the accident.

17.INJURY_CLASSIFICATION: The severity of injuries sustained by the person (e.g., fatal, serious, minor, none).

18.HOSPITAL: The hospital where the injured person was taken after the accident.

19.EMS_AGENCY: The emergency medical service agency that responded to the accident.

20.EMS_RUN_NO: The run number assigned by the EMS agency for this particular incident.

21.DRIVER_ACTION: The action taken by the driver that might have contributed to the accident (e.g., speeding, distracted driving, running a red light).

22.DRIVER_VISION: Conditions affecting the driver’s vision at the time of the accident (e.g., glare, obstruction, weather conditions).

23.PHYSICAL_CONDITION: The physical condition of the driver (e.g., sober, under the influence, fatigued).

24.PEDPEDAL_ACTION: Actions of the pedestrian or pedalcyclist that might have contributed to the accident (e.g., crossing road, not visible).

25.PEDPEDAL_VISIBILITY: Visibility conditions of the pedestrian or pedalcyclist (e.g., wearing reflective clothing, poor visibility).

26.PEDPEDAL_LOCATION: The location of the pedestrian or pedalcyclist at the time of the accident (e.g., crosswalk, roadway, sidewalk).

27.BAC_RESULT: Blood Alcohol Concentration (BAC) test result for the driver.

28.BAC_RESULT_VALUE: The numerical value of the BAC test result.

29.CELL_PHONE_USE: Indicates whether the driver was using a cell phone at the time of the accident.
