```python
def remove_outliers_zscore(df):
    """
    Takes in a dataframe and removes outliers using the Z-score method.
    """
    import numpy as np
    
    # Convert columns to numeric (if they are not already)
    df = df.apply(pd.to_numeric, errors='coerce')
    
    # Calculate Z-scores
    z_scores = np.abs((df - df.mean()) / df.std())
    
    # Identify outliers
    outliers_z = df[(z_scores > 3).any(axis=1)]
    
    # Remove outliers from dataframe
    df_cleaned = df.drop(outliers_z.index)
    
    return df_cleaned
```

```python
# recheck for outliers
# Load well_features.csv into a DataFrame
df = pd.read_csv("well_features.csv")

# Call the function to remove outliers
df_cleaned = remove_outliers_zscore(df)

# Check if there are any outliers remaining
outliers_check = df.shape[0] - df_cleaned.shape[0]
if outliers_check > 0:
    print(f"There were {outliers_check} outliers removed.")
else:
    print("No outliers were found.")
```
